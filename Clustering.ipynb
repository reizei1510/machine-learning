```python
#8
from sklearn.cluster import KMeans

X = df[['Spending Score (1-100)', 'Age']]
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¾Ð²')
plt.ylabel('Ð˜Ð½ÐµÑ€Ñ†Ð¸Ñ')
plt.title('ÐœÐµÑ‚Ð¾Ð´ Ð»Ð¾ÐºÑ‚Ñ Ð´Ð»Ñ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ñ Ð¾Ð¿Ñ‚Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ð³Ð¾ Ñ‡Ð¸ÑÐ»Ð° ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¾Ð²')
plt.show()
```
