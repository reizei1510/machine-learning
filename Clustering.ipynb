```python
#8
from sklearn.cluster import KMeans

X = df[['Spending Score (1-100)', 'Age']]
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')
plt.title('Метод локтя для определения оптимального числа кластеров')
plt.show()
```

```python
#9
from sklearn.decomposition import PCA

optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Визуализация результатов кластеризации с использованием метода главных компонент
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Создание DataFrame для удобства визуализации
cl = pd.DataFrame({'PC1': X_pca[:, 0], 'PC2': X_pca[:, 1], 'Cluster': labels})

# Визуализация результатов кластеризации
plt.figure(figsize=(8, 6))
cl.plot.scatter(x='PC1', y='PC2')
plt.title('Результаты кластеризации с KMeans (PCA)')
plt.show()
```
